```python
import torch
import torchvision.utils as vutils  
from torch.autograd import Variable

import sys
import json
import numpy
import models.gan as dcgan
import matplotlib.pyplot as plt
import math

import random

modelToLoad = "/home/hemanth/Downloads/lode_runner/GAN_pytorch/model_epoch_1000_0_10.pth"

batchSize = 1
nz = 10# Dimensionality of latent vector

imageSize = 32
ngf = 64
ngpu = 1
n_extra_layers = 0
z_dims = 6  # number different titles

generator = dcgan.GanGenerator(
     imageSize, nz, z_dims, ngf, ngpu, n_extra_layers)
 #generator.load_state_dict(torch.load('netG_epoch_24.pth', map_location=lambda storage, loc: storage))
generator.load_state_dict(torch.load(
     modelToLoad, map_location=lambda storage, loc: storage), strict=False)

```

```python
line = []
for i in range(batchSize):
    line.append([random.uniform(0.0, 1.0)]*nz)
   #This is the format that we expect from sys.stdin
line = json.dumps(line)
lv = numpy.array(json.loads(line))
 # torch.from_numpy(lv)# torch.FloatTensor( torch.from_numpy(lv) )
latent_vector = torch.Tensor(lv).view(batchSize, nz, 1, 1)
```

```python
hl_graph = hl.build_graph(generator, latent_vector)
hl_graph.theme = hl.graph.THEMES["blue"].copy()  # Two options: basic and blue
hl_graph
```
