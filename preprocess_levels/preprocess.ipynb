```python
import os,glob
levels_list=[]
levels=[]
folder_path = '/home/hemanth/Downloads/lode_runner/preprocess_levels'
for filename in glob.glob(os.path.join(folder_path, '*.txt')):
  with open(filename, 'r') as f:
      levels_list.append(''.join(f.read().split()))
print(levels_list[0])
print(len(levels_list))
```

```python
print(levels_list[0])
for i in range(0,len(levels_list)):
    levels_list[i] = levels_list[i].translate(str.maketrans({'.':'0','E':'0','G':'1','b':'2','B':'3','#':'4','M':'0','-':'5'}))
levels_list[0]
```

```python
import numpy as np
levels = []
for i in levels_list:
    levels.append(list(map(int,i)))
levels_np = np.array(levels)
levels_np = levels_np.reshape(150,22,32)
print(levels_np.shape)
levels_np[0]
```

```python
import json
with open('/home/hemanth/Downloads/lode_runner/GAN_pytorch/levels.json', 'w') as outfile:
    json.dump(levels_np.tolist(), outfile)
```
